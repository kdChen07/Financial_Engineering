```python
import sympy as sp
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')      # 文件读取要确定好位置
df.insert(1, '天数', range(1,len(df)+1))
```

```python
df1=df[1:10]
```

```python
def MLE(df1):
    u,q = sp.symbols('u q')     # u,q表示均值和方差
    t = df['天数']
    s = df['TTF']
    s0 = df.loc[450,'TTF']
    F = 1
    for i in range(1,len(df1)):
        f = sp.exp( -(np.log10(s[i]) - np.log10(s0) - (u - 0.5* q**2)* t[i])**2 / (2* q**2 * t[i]) ) / (np.sqrt(2 * np.pi * t[i]) * q * s[i])
        F = F * f
    logF = sp.expand_log(sp.log(F))
    logF_u = sp.diff(logF,u)
    logF_q = sp.diff(logF,q)
    result = sp.solve([logF_u, logF_q],[u, q])
    return result
result = MLE(df1)
print(result)
```
