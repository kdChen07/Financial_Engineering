```python
import sympy as sp
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('data2.csv')      # 文件读取要确定好位置
df.insert(1, '天数', range(1,len(df)+1))
```

```python
def MLE(df1):
    a,q,w = sp.symbols('a q w')     # u,q表示均值和方差
    t = df['天数']
    s = df['TTF']
    # s0 = df.loc[450,'TTF']
    F = 0
    for i in range(1,len(df1)):
        f = sp.exp(( -(s[i]-sp.exp(-a*1)*s[i-1]-(w/a)*(1-sp.exp(-a*1)))**2)/((s[i-1]*(q**2))/a)*(1-sp.exp(-2*a*1)))/((sp.sqrt(np.pi*(q**2)*s[i-1]))/a)*(1-sp.exp(-2*a*1))
        # f = sp.exp( -(np.log10(s[i]) - np.log10(s0) - (u - 0.5* q**2)* t[i])**2 / (2* q**2 * t[i]) ) / (np.sqrt(2 * np.pi * t[i]) * q * s[i])
        F = F + sp.log(f)
    logF = sp.expand_log(F)
    logF_a = sp.diff(logF,a)
    logF_q = sp.diff(logF,q)
    logF_w = sp.diff(logF,w)
    result = sp.solve([logF_a, logF_q, logF_w],[a, q, w])
    return result
result = MLE(df1)
print(result)
```
