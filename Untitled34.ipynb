```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
N1=10000
N2=250
S=100
b=np.ones(N1)
S0=S*b
T=365/365
t_delta=T/N2
mu=0.05
r=0.15
sigma=0.4
K=100
a=2
```

```python
St=np.zeros((N1,N2))
St[:,0]=S0
St
```

```python
sum=0
for i in range(0,N1):
  Epsilon = np.random.normal(0, 1, N2)
  for t in range(1,N2):
    St[i,t]=St[i,t-1]+a*(100-St[i,t-1])*t_delta+sigma*math.sqrt(St[i,t-1])*math.sqrt(t_delta)*Epsilon[t]
```

```python
St
```

```python
x = np.arange(0,N2,1)
y = St
```

```python
plt.subplots(figsize=(16,9),dpi=100)
for i in range(0,N2):
  plt.plot(x, y[i,:])
  plt.title('Monte Carlo')
  plt.xlabel('Step')
  plt.ylabel('Price')
```

```python
plt.subplots(figsize=(16,9),dpi=100)
for i in range(0,2):
  plt.plot(x, y[i,:])
  plt.title('Monte Carlo')
  plt.xlabel('Step')
  plt.ylabel('Price')
```
