```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
N1=100000
N2=100
S=30
a=np.ones(N1)
S0=S*a
Y0=0.015
Y=Y0*a
XY=0.01
sigma_Y=0.05
aY=2
r0=0.05
r=r0*a
Xr=0.05
sigma_r=0.05
T=365/365
t_delta=T/N2
mu=0.1
r=0.1
sigma=0.3
K=30
```

```python
Y=np.zeros((N1,N2))
Y[:,0]=Y0
Y
```

```python
for i in range(0,N1):
  Epsilon = np.random.normal(0, 1, N2)
  for t in range(1,N2):
    Y[i,t]=Y[i,t-1]+aY*(XY-max(Y[i,t-1],0))*t_delta+sigma_Y*math.sqrt(max(Y[i,t-1],0))*math.sqrt(t_delta)*Epsilon[t]
```

```python
r=np.zeros((N1,N2))
r[:,0]=r0
r
```

```python
x = np.arange(0,N2,1)
y = Y
```

```python
plt.subplots(figsize=(32,9),dpi=100)
for i in range(0,N2):
  plt.plot(x, y[i,:])
  plt.title('Monte Carlo')
  plt.xlabel('Step')
  plt.ylabel('Price')
  plt.axhline(0.01, color='r', ls='dashed', lw=1.5,
            label='均值')
```
